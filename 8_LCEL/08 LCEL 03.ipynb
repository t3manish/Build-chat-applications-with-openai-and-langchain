```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
```

```python
chat_template = ChatPromptTemplate.from_messages([
    ('human',
     "I've recently adopted a {pet} which is a {breed}. Could you suggest several training tips?")])
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  model_kwargs = {'seed':365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
chain = chat_template | chat
```

```python
chain.invoke({'pet':'dog', 'breed':'shepherd'})
```

```python
%%time
chain.batch([{'pet':'dog', 'breed':'shepherd'},
             {'pet':'dragon', 'breed':'night fury'}])
```

```python
%%time
chain.invoke({'pet':'dog', 'breed':'shepherd'})
```

```python
%%time
chain.invoke({'pet':'dragon', 'breed':'night fury'})
```

Streaming allows you to receive the response from the language model token by token, rather than waiting for the entire response to be generated. This is particularly useful for interactive applications, as it provides a more responsive user experience. You can see the text appearing as it's generated, rather than after the full generation is complete. The existing cells for streaming didn't correctly demonstrate this. I will fix them to show how to iterate over the streamed response.

```python
response = chain.stream({'pet':'dragon', 'breed':'night fury'})
```
